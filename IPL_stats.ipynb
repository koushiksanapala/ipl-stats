```
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```
#loads data
matches = pd.read_csv('Match_Info.csv')
teams = pd.read_csv('teams_info.csv')
print(matches.head())
print(teams.head())
```

```
print("match info columns:", matches.columns)
print("team info columns", teams.columns)
```

```
matches['winner'].value_counts().plot(kind='bar', title='Most Match Wins', color='purple')
```

```
import matplotlib.pyplot as plt

# Step 1: Count number of wins by each team
team_wins = matches['winner'].value_counts()

# Step 2: Plot it
team_wins.plot(kind='bar', color='purple')
plt.title('Most Match Wins')
plt.xlabel('Team')
plt.ylabel('Wins')
plt.xticks(rotation=45)
plt.show()
```

```
matches['winner'].value_counts().plot(kind='pie',autopct='%d%%',figsize=(8,8),startangle=140,colormap='plasma',legend=True)
plt.title("win % by each team")
plt.ylabel("teams")
plt.show()
```

```
matches['winner'].value_counts().plot(kind='barh', color='skyblue')
plt.title('Most Match Wins (Horizontal View)')
plt.xlabel('Wins')
plt.ylabel('Teams')
plt.show()
```

```
remove_teams = [
    'Kochi Tuskers Kerala',
    'Pune Warriors',
    'Rising Pune Supergiants',
    'Rising Pune Supergiant',
    'Gujarat Lions'
    'Deccan Chargers'
]
```

```
filtered_matches = matches[~matches['winner'].isin(remove_teams)]
```

```
filtered_matches['winner'].value_counts().plot(
    kind='bar',
    color='green',
    title='Filtered Match Wins (Active Teams Only)'
)
```
